* A t elle un sens ? dans le cas des fluctions d'un profil soumis à l'agitation thermique
* De base ça marche pour des états discrets (pour la th des éléctrons dans le solide)

```python
# function [ndos,x_ndos]=calc_dos(ND,E,Nt)
    
#     Emax=E(Nt)
#     Emin=E(1)
#     ndos=zeros(ND,1);
    
#     delta_D=(Emax-Emin)/ND // calcul de la valeur de l'intervalle
#     for k=1:ND, // Boucle sur chaque intervalle
#     x_ndos(k)=Emin+(k-0.5)*delta_D;
#     for i=1:Nt,
#        if E(i)>=Emin+(k-1)*delta_D then 
#        if E(i)<Emin+k*delta_D then ndos(k)=ndos(k)+1; 
#            end; end; end;
#     end // Fin boucle sur chaque intervalle
#     ndos=ndos/(delta_D*Nt); // normalisation n(E)=dN(E)/dE
# endfunction
def calc_dos(ND,E,Nt):
    Emax = max(E)
    Emin = min(E)
#     ndos=np.zeros((ND,1))
#     x_ndos=np.zeros((ND,1))
    ndos=[]
    x_ndos=[]
    delta_D = (Emax-Emin)/ND
    for k in range(ND):# // Boucle sur chaque intervalle
        x_ndos.append(Emin+(k-0.5)*delta_D)
    for i in range(Nt):#
        if(E[i]>=Emin+(k-1)*delta_D):
            ndos.append( 1)
        elif (E[i] < Emin+k*delta_D):
                ndos.append(1)
    ndos=ndos/(delta_D*Nt) #normalization n(E)=dN(E)/dE

    return(ndos,x_ndos)
ND= len(w)
Nt= len(w) #dimension totale 
(ydos,xdos)=calc_dos(ND,w,Nt) #// deuxième façon de calculer la dos


#/ Fonction qui calcule la densité d’états à partir des niveau d’énergie
# En entrée : 
# E(i) : Le tableau des énergies
# eps : une valeur de coupure (très petite) au cas où les niveau soient dégénérés

def DOS_ener(E,nE,eps):
    delt=[]
    xdos=[]
    ydos=[]
    for  i in range(nE-1):
        delt.append(E[i+1]-E[i]);
        xdos.append(E[i]);
        ydos.append((1./(delt[i]-i*eps)).real)
    return(xdos,ydos)

eps=1e-5
E=w
nE=len(w)

(ydos2,xdos2)=DOS_ener(E,nE,eps)

# plt.figure(figsize=(10, 12))
# plt.plot(xdos,ydos)
plt.figure(figsize=(10, 12))
plt.plot(ydos2,xdos2)
```

```python
#Adapted from https://www.lfd.uci.edu/~gohlke/ipcf/
def dot_python(a, b, start, stop, delay):
    """Return dot product of two sequences in range."""
    sum = 0
    for n in range(start, stop):
        sum += a[n + delay] * b[n]
    return sum /stop


def correlate_python(a, b):
    """Return linear correlation of two sequences."""
    size = len(a)

    c = [0] * size  # allocate output array/list

    for index in range(size):
        #delay =np.asarray(index - size // 2)*1e-8
        delay =index - size // 2
        if delay < 0:
            c[index] = dot_python(a, b, -delay, size, delay)
        else:
            c[index] = dot_python(a, b, 0, size-delay, delay)

    return c
```

```python
def sf_from_positions(positions,box_dimension):

    sf = list()
    modes = list(x for x in range(1,int(box_dimension)))

    dk = 2*np.pi/box_dimension

    for h in modes:

        k_vec = np.array([1,1])*dk*h
        summation = 0

        for position in positions:

            summation += np.exp(-1j*k_vec.dot(position))

        sf.append(abs(summation)**2/len(positions))

    return sf, modes

rows,columns = np.nonzero(np.asarray(content[-1]))
positions = list(zip(rows,columns))

factor, wave_modes = sf_from_positions(positions,dimension)


# rows = shape(content[-1][0])
# columns = 0
# positions = list(zip(rows,columns))
#dimension = np.shape(triangular_lattice)[0]
L = dx*h0*len(content[-1])
dimension = np.int(L/lambda_mean)
factor, wave_modes = sf_from_positions(positions,dimension)
```
